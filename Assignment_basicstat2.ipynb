### Problem Statement : Hospital Patient Data Analysis


Context: A hospital maintains patient records including admission details, department, diagnosis, doctor, and bill amount. You have two datasets: one with patient info and another with billing details. Some patients have blank bill amounts, and there are multiple rows for the same patient due to follow-ups.

Tasks:

1. Load the patient dataset and show summary with info().

2. Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].

3. Drop administrative columns like ['ReceptionistID', 'CheckInTime'].

4. Use groupby to find total bill amount per department.

5. Remove duplicate patient records based on PatientID.

6. Fill missing BillAmount values with the mean bill amount.

7. Merge the billing dataset with patient dataset on PatientID.

8. Concatenate an additional DataFrame that contains new patients for the current week (row-wise).

9. Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Patient_Data.csv') #Loading the patient dataset
```

```python
df.info() #show summary
```

```python
df
```

```python
df1 = ['PatientID','Name','Department','Doctor','BillAmount'] #Select only the columns relevant for billing
```

```python
df2 = df[df1] 
```

```python
df2
```

```python
df = df.drop(columns = ['ReceptionistID', 'CheckInTime']) #Drop administrative columns
```

```python
df
```

```python
df.groupby('Department')['BillAmount'].sum().reset_index() #Use groupby to find total bill amount per department.
```

```python
df.drop_duplicates(subset='PatientID',inplace= True) #Remove duplicate patient records based on PatientID.
```

```python
df.isnull().sum()
```

```python
df.fillna({'BillAmount':df.BillAmount.mean()}, inplace=True) #Fill missing BillAmount values with the mean bill amount.
```

```python
df
```

```python
bill_df = pd.read_csv('Billing_Data.csv')
```

```python
bill_df
```

```python
pd.merge(df,bill_df) #Merge the billing dataset with patient dataset on PatientID.
```

```python
dict = ({
    'PatientID': [201, 202, 203],
    'Name': ['Neha', 'Ravi', 'Karan'],
    'Department': ['Cardiology', 'Neurology', 'Orthopedics'],
    'Doctor': ['Dr. Mehta', 'Dr. Rao', 'Dr. Singh'],
    'BillAmount': [15000, 20000, 12000]
})
```

```python
new_patients_df = pd.DataFrame(dict) 
```

```python
new_patients_df
```

```python
df = pd.concat([df,new_patients_df]) ##Concatenate an additional DataFrame that contains new patients
```

```python
df
```

```python
new_billing_cols = pd.DataFrame({
    'InsuranceCovered': [2000.0, 1500.0, 2500.0, 3000.0, 1000.0],
    'FinalAmount': [3000.0, 3500.0, 5000.0, 3200.0, 4000.0]
})
```

```python
new_billing_cols
```

```python
df = pd.concat([df, new_billing_cols], axis=1)
```
